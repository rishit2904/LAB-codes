## Question 2

```python
x = [1, 2, 3]
y = [1, 4, 9]

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])      
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  

ax1.plot(x, y)  
ax2.plot(x, y, 'r')

ax1.set_xlabel('X Label for ax1')
ax1.set_ylabel('Y Label for ax1')
ax1.set_title('Title for ax1')

ax2.set_xlabel('X Label for ax2')
ax2.set_ylabel('Y Label for ax2')
ax2.set_title('Title for ax2')

plt.show()
```
